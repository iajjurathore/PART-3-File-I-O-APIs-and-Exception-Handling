##STEP 1 — FETCH AND DISPLAY 20 PRODUCTS##

```python
import requests

BASE_URL = "https://dummyjson.com/products"
# 
##STEP 1 — FETCH AND DISPLAY 20 PRODUCTS##

response = requests.get(f"{BASE_URL}?limit=20")
data = response.json()
products = data["products"]
print("=" * 80)
print("STEP 1: All 20 Products")
print("=" * 80)
print(f"{'ID':<5} | {'Title':<30} | {'Category':<20} | {'Price':>10} | {'Rating':>8}")
print("-" * 80)
for p in products:
    print(f"{p['id']:<5} | {p['title']:<30} | {p['category']:<20} | ${p['price']:>9.2f} | {p['rating']:>8.2f}")
```

#STEP 2 - FILTER & SORT ##

```python
#STEP 2 - FILTER & SORT ##

high_rated = [p for p in products if p["rating"] >= 4.5]
sorted_products = sorted(high_rated, key=lambda p: p["price"], reverse=True)
print("\n" + "=" * 80)
print("STEP 2: High-Rated Products (Rating >= 4.5), Sorted by Price Descending")
print("=" * 80)
print(f"{'ID':<5} | {'Title':<30} | {'Category':<20} | {'Price':>10} | {'Rating':>8}")
print()
for p in sorted_products:
    print(f"{p['id']:<5} | {p['title']:<30} | {p['category']:<20} | ${p['price']:>9.2f} | {p['rating']:>8.2f}")
print(f"Total: {len(sorted_products)} products")
```

##STEP 3 - SEARCH BY CATEGORY## 

```python
##STEP 3 - SEARCH BY CATEGORY## 

laptops_response = requests.get(f"{BASE_URL}/category/laptops")
laptops = laptops_response.json()["products"]
print("\n" + "=" * 80)
print("STEP 3: Laptops Category")
print("=" * 80)
print(f"{'Title':<40} | {'Price':>10}")
print("-" * 55)
for laptop in laptops:
    print(f"{laptop['title']:<40} | ${laptop['price']:>9.2f}")
print(f"Total laptops found: {len(laptops)}")
```

##STEP 4 — POST REQUEST ##

```python
## STEP 4 — POST REQUEST ##
 
new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}
post_response = requests.post(f"{BASE_URL}/add", json=new_product)
post_data = post_response.json()
print("\n" + "=" * 80)
print("STEP 4: POST Response (Simulated)")
print("=" * 80)
print(f"Status Code: {post_response.status_code}")
for key, value in post_data.items():
    print(f"  {key:<15}: {value}")
```

##PART A — GUARDED CALCULATOR##

```python
import requests

## PART A — GUARDED CALCULATOR##

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"
print("=== Part A: Safe Divide ===")
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))
```

##PART B — GUARDED FILE READER ##

```python
## PART B — GUARDED FILE READER ##

def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            content = f.read()
        return content
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return None
    finally:
        print("File operation attempt complete.")
print("\n=== Part B: File Reader ===")
read_file_safe("Python_notes.txt")
read_file_safe("ghost_file.txt")
```

#PART C — ROBUST API HELPERS ##

```python
## PART C — ROBUST API HELPERS ##
 
def safe_get(url, timeout=5):
    try:
        return requests.get(url, timeout=timeout)
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")
    return None
def safe_post(url, json_data, timeout=5):
    try:
        return requests.post(url, json=json_data, timeout=timeout)
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")
    return None
```

##PART D — INPUT VALIDATION LOOP ##

```python
## PART D — INPUT VALIDATION LOOP ##

print("\n=== Part D: Product Lookup ===")
while True:
    user_input = input("\nEnter a product ID (1-100), or 'quit' to exit: ").strip()
    if user_input.lower() == "quit":
        print("Goodbye!")
        break
    try:
        product_id = int(user_input)
    except ValueError:
        print("Warning: Please enter a valid integer between 1 and 100.")
        continue
    if not (1 <= product_id <= 100):
        print("Warning: Product ID must be between 1 and 100. Try again.")
        continue
    response = safe_get(f"https://dummyjson.com/products/{product_id}")
    if response is None:
        continue
    if response.status_code == 404:
        print("Product not found.")
    elif response.status_code == 200:
        product = response.json()
        print(f"Title : {product['title']}")
        print(f"Price : ${product['price']}")
    else:
        print(f"Unexpected status code: {response.status_code}")
```

==
Task 4 — Logging to File
==

```python
import requests
from datetime import datetime
LOG_FILE = "error_log.txt"

## Logger function ##

def log_error(function_name, error_type, message):
    """Append a timestamped error entry to error_log.txt."""
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    entry = f"[{timestamp}] ERROR in {function_name}: {error_type} - {message}\n"
    with open(LOG_FILE, "a", encoding="utf-8") as f:
        f.write(entry)
    print(f"Logged: {entry.strip()}")





## Entry 1: Trigger ConnectionError ##
 
def fetch_products():
    url = "https://this-host-does-not-exist-xyz.com/api"
    try:
        response = requests.get(url, timeout=5)
        return response.json()
    except requests.exceptions.ConnectionError as e:
        log_error("fetch_products", "ConnectionError", str(e)[:80])
        return None
    except requests.exceptions.Timeout:
        log_error("fetch_products", "Timeout", "Request timed out after 5 seconds")
        return None
    except Exception as e:
        log_error("fetch_products", type(e).__name__, str(e)[:80])
        return None




## Entry 2: Trigger HTTP  ##

def lookup_product(product_id):
    url = f"https://dummyjson.com/products/{product_id}"
    try:
        response = requests.get(url, timeout=5)
        if response.status_code != 200:
            log_error(
                "lookup_product",
                "HTTPError",
                f"{response.status_code} Not Found for product ID {product_id}"
            )
            return None
        return response.json()
    except requests.exceptions.ConnectionError as e:
        log_error("lookup_product", "ConnectionError", str(e)[:80])
        return None
    except requests.exceptions.Timeout:
        log_error("lookup_product", "Timeout", "Request timed out after 5 seconds")
        return None
    except Exception as e:
        log_error("lookup_product", type(e).__name__, str(e)[:80])
        return None
 
```

##Run and read log ##

```python

## Run and read log ##

print("=== Triggering ConnectionError ===")
fetch_products()
print("\n=== Triggering HTTP 404 Error ===")
lookup_product(999)
print("\n=== Contents of error_log.txt ===")
with open(LOG_FILE, "r", encoding="utf-8") as f:
    print(f.read())
```
